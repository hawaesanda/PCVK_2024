```python
import numpy as mp
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

#Membuat list untuk menyimpa url dari beberapa image
urls = ["https://i.pinimg.com/564x/2a/44/8a/2a448a2465f4eb537f89a75dc62d0595.jpg",
        "https://i.pinimg.com/564x/87/e2/f0/87e2f0e3ab3a2f719da1ee3db638d71e.jpg",
        "https://i.pinimg.com/564x/f1/a1/a2/f1a1a28d95b667fff10e5b10e615a00f.jpg"]
#baca dan tampilkan image
#loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print("\n")
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  # dengan loop
  # image_3[int((tinggi)/2), y] = [255, 255, 255]
  # dengan tanpa loop
  image_3[int((tinggi)/2), :] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal dari kiri atas ke kanan bawah
for x in range(tinggi):
  y = int(x * lebar / tinggi)
  if y < lebar:
    image_5[x, y] = [255, 255, 255]

# Membuat garis diagonal dari kanan atas ke kiri bawah
for x in range(tinggi):
  y = lebar - int(x * lebar / tinggi) - 1  # Menghitung posisi y untuk diagonal kedua
  if y >= 0:
    image_5[x, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

```python
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Tentukan panjang garis horizontal
panjang_garis = int(lebar * 0.6)  # Misalnya, 60% dari lebar gambar

# Tentukan titik awal garis (tengah gambar dikurangi setengah panjang garis)
x_start = int(lebar / 2 - panjang_garis / 2)

# Gambar garis horizontal
for x in range(x_start, x_start + panjang_garis):
  image_6[int(tinggi / 2), x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
```

```python
import cv2
import numpy as np
from skimage import io
from google.colab.patches import cv2_imshow  # import cv2_imshow from google.colab.patches

# Define the URL of the image
url = "https://umsu.ac.id/artikel/wp-content/uploads/2023/11/cara-mudah-cek-ktp-asli-atau-palsu.jpeg"
# Load the image from the URL
image = io.imread(url)

# Convert image from RGB (skimage) to BGR (OpenCV)
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

# Convert the image from BGR to RGB for image_2 and image_3
image_2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Get image dimensions
tinggi, lebar = image.shape[:2]

# Define the coordinates of the box
x1, y1 = 100, 20  # Top-left corner
x2, y2 = 175, 710  # Bottom-right corner

# Fill the box with white color
for x in range(x1, x2):
    for y in range(y1, y2):
        if 0 <= x < tinggi and 0 <= y < lebar:  # Ensure coordinates are within the image bounds
            image_3[x, y] = [0, 0, 255]  # White color (in RGB)

# Combine the two images horizontally
final_frame = cv2.hconcat([image_2, image_3])

# Display the result
cv2_imshow(final_frame)  # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```
